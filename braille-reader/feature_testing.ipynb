# Braille Reader Feature Testing

```python
import os
from PIL import Image
import matplotlib.pyplot as plt
from src.image_processor import ImageProcessor
from src.converter import Converter
from src.visualizer import Visualizer
from src.document_processor import DocumentProcessor
from src.braille_translator import BrailleTranslator

# Initialize classes
image_processor = ImageProcessor()
converter = Converter(image_processor)
visualizer = Visualizer()
document_processor = DocumentProcessor()
braille_translator = BrailleTranslator()
```

## Load an Image

```python
image_path = '''.\examples\Screenshot 2025-07-10 225514.png'''  # Example image path
image_processor.load_image(image_path)
original_image = image_processor.image
plt.imshow(original_image)
plt.title('Original Image')
plt.axis('off')
plt.show()
```

## Convert to Black and White

```python
image_processor.convert_to_bw()
bw_image = image_processor.image
plt.imshow(bw_image, cmap='gray')
plt.title('Black and White Image')
plt.axis('off')
plt.show()
```

## Resize Image

```python
new_width, new_height = 200, 300  # Example dimensions
image_processor.resize_image(new_width, new_height)
resized_image = image_processor.image
plt.imshow(resized_image, cmap='gray')
plt.title('Resized Image')
plt.axis('off')
plt.show()
```

## Convert Image to Matrix

```python
matrix_representation = converter.image_to_matrix()
print(matrix_representation)
```

## Load a Document

```python
document_path = '''.\examples\pdf.pdf'''  # Example document path
document_processor.load_document(document_path)
extracted_text = document_processor.text
print(extracted_text)
```

## Translate Text to Braille

```python
braille_representation = braille_translator.text_to_braille(extracted_text)
print(braille_representation)
```

## Visualize Original and Converted Images

```python
visualizer.display_images([original_image, bw_image, resized_image], titles=['Original', 'Black & White', 'Resized'])
```

```python
# Convert document text to image
text_image = document_processor.text_to_image()
# Resize the text image to lower pixel dimensions
resized_text_image = document_processor.resize_image(200, 300)

# Visualize both images side by side
visualizer.display_images(
    [text_image, resized_text_image],
    titles=['Document as Image', 'Resized Document Image']
)
```
